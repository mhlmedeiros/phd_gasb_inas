```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.axes_grid1.inset_locator import (inset_axes, InsetPosition, mark_inset)
import numpy as np

FONT_LABELS = 18
FONT_TITLES = 20
font = {'family' : 'serif', 'weight' : 'bold', 'size': FONT_LABELS}
matplotlib.rc('font', **font)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
transport_file_1 = "data_439.7_440_meV_Fermi_Transport_Total_1001.npy"
fermi_energy_1   = "data_439.7_440_meV_Fermi_1001.npy"
trans_1  = np.load(transport_file_1)
energy_1 = np.load(fermi_energy_1)

transport_file_2 = "data_439.7_440_meV_Fermi_Transport_Total_1001_L_707.1067811865476.npy"
fermi_energy_2   = "data_439.7_440_meV_Fermi_1001_L_707.1067811865476.npy"
trans_2  = np.load(transport_file_2)
energy_2 = np.load(fermi_energy_2)

transport_file_3 = "data_439.7_440_meV_Fermi_Transport_Total_1001_L_1000.npy"
fermi_energy_3   = "data_439.7_440_meV_Fermi_1001_L_1000.npy"
trans_3  = np.load(transport_file_3)
energy_3 = np.load(fermi_energy_3)

extrema_file_1 = "data_439.7_440_meV_Fermi_Transport_Total_1001_L_707.1067811865476_extrema_Fermi_439_440.npy"
extrema_file_fermi_energy_1 = "data_439.7_440_meV_Fermi_1001_L_707.1067811865476_extrema_Fermi_439_440.npy"
extrema_val_1  = np.load(extrema_file_1)
extrema_val_energy_1 = np.load(extrema_file_fermi_energy_1)

extrema_file_2 = "data_439.7_440_meV_Fermi_Transport_Total_1001_L_1000_extrema_Fermi_439_440.npy"
extrema_file_fermi_energy_2 = "data_439.7_440_meV_Fermi_1001_L_1000_extrema_Fermi_439_440.npy"
extrema_val_2  = np.load(extrema_file_2)
extrema_val_energy_2 = np.load(extrema_file_fermi_energy_2)
```

```python
figure, ax1 = plt.subplots(figsize=(9,6))

ax1.plot(energy_1, trans_1, 
         #marker = 'o',
         #markevery = 10,
         linewidth = 2,
         color = 'r',
         label = r'$L = 500$ nm')
ax1.plot(energy_2, trans_2, 
         #marker = 's',
         linestyle = '--',
         linewidth = 0.8,
         color = 'b',
         #markevery = 10,
         label = r'$L = 500\sqrt{2}$ nm')
ax1.plot(energy_3, trans_3, 
         #marker = '^',
         linestyle = '-.',
         linewidth = 0.8,
         color = 'g',
         #markevery = 10, 
         label = r'$L = 1~\mu$m')

ax1.plot(extrema_val_energy_1, extrema_val_1, 
         marker = '^',
         linestyle = '',
#          linewidth = 0.8,
         color = 'b')
         #markevery = 10, 
#          label = r'$L = 500\sqrt{2}$ nm extr.')

ax1.plot(extrema_val_energy_2, extrema_val_2, 
         marker = '^',
         linestyle = '',
#          linewidth = 0.8,
         color = 'g')
         #markevery = 10, 
#          label = r'$L = 1~\mu$m extr.')


ax1.legend()
ax1.set_xlim(440,439.75)
plt.tight_layout()
plt.savefig("comparison_transport_length.png")
plt.show()
```
