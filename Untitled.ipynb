```python
def prepara_array(dados):

    dados = dados[1:-1]

    x_list = dados.split(" ")
    x_list = [el.rstrip('\n') for el in x_list]

    x_new = []
    for el in x_list:
        if el != '':
            x_new.append(float(el))
    return x_new
```
